```python
import seaborn as sns
import pandas as pd

# load the iris dataset
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')

# create a violin plot of sepal length
sns.violinplot(data=iris, y="sepal_length")
```

```python
# Load Tips Dataset from Seaborn
tips = sns.load_dataset("tips")
tips
```

```python
# Scatter Plot

sns.scatterplot(x=tips.total_bill, y=tips.tip, data=tips, hue="smoker")
```

```python
tips.smoker.value_counts()
```

```python
sns.relplot(x=tips.total_bill, y=tips.tip, data=tips, hue="smoker")
```

```python
tips["size"].value_counts()
```

```python
sns.relplot(x=tips.total_bill, y=tips.tip, data=tips, hue="size")
```

```python
sns.relplot(x=tips.total_bill, y=tips.tip, data=tips, size="size", hue="smoker")
```

```python
sns.relplot(x=tips.total_bill, y=tips.tip, data=tips, style="size", hue="smoker")
```

```python
tips["day"].value_counts()
```

```python
# Categorical Plot (to show ploting on Categorical data vs numerical data)
sns.catplot(x= 'smoker', y='tip', data=tips)
```

```python
sns.catplot(x= 'day', y='total_bill', data=tips)
```

```python
# Joinplot (combination of scatter plot and histogram)
sns.jointplot(x=tips.total_bill, y=tips.tip)
```

```python
# Pait Plor (ALl combinations of plots)
sns.pairplot(tips)
```

```python
sns.pairplot(iris)
```
