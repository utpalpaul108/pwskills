## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(20)
y = np.random.rand(20)
```

```python
x
```

```python
y
```

```python
## Draw Scatter Plot
plt.figure(figsize=(6,2)) # Control the size of the plot
plt.scatter(x,y, c="b", alpha=0.8, marker="*") # 'c' is for color, 'alpha' is for color intensity, 'marker' is for dot shape
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Scatter Plot of X vs Y")
plt.grid() # Add grid
```

```python
x = np.linspace(1,10, 100)
y = np.sin(x)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y, "--g", )
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Sin Curve")
# plt.grid()
```

```python
# Bar Plot
x = ["A", "B", "C", "D", "E"]
y = np.random.rand(5)
```

```python
plt.figure(figsize=(4,3))
plt.bar(x,y,color="b")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("X Vs Y Bar Chart")
plt.grid()
```

```python
plt.figure(figsize=(8,3))
plt.barh(x,y,color="b")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("X Vs Y Bar Chart")
plt.grid()
```

```python
x = [2,4,5,3,3,5,2,2,5,6,3,4,3]
plt.figure(figsize=(6,2))
plt.hist(x, color="b")
plt.xlabel("Number")
plt.ylabel("Frquency")
plt.title("Number Vs Frequency")
plt.grid()
plt.savefig("hist.png")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.scatter(x,y,z,c="b")
plt.show()
```
