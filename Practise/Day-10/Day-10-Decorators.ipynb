## Decorators

```python
def my_dec(func):
    def my_inner_deco(msg):
        print("Start of the function")
        func(msg)
        print("End of the function")
    
    return my_inner_deco
```

```python
@my_dec
def my_func(message):
    print(message)
```

```python
my_func("Hello World")
```

```python
def add(x):
    return x+1

def sub(x):
    return x-1
    
def my_func(func, x):
    x = func(x)
    return x
```

```python
my_func(add,10)
```

```python
my_func(sub,10)
```

```python
def hello():
    def hi():
        return "Hello Everyone"
    def hy():
        return "Hy Everyone"
    return hi
```

```python
x = hello()
x()
```

```python
def devide(x,y):
    print(x/y)

def outer_func(func):
    def inner_func(x,y):
        if x<y:
            x,y = y,x
        func(x,y)
    return inner_func
```

```python
my_func = outer_func(devide)
```

```python
x = my_func(2,10)
x
```

```python
@outer_func
def devide(x,y):
    print(x/y)
```

```python
x = devide(2,10)
x
```

```python
def outer_twice(func):
    def inner_twice():
        print("Start of the function")
        func()
        func()
        print("End of the Function")
    return inner_twice
```

```python
@outer_twice

def gretting(*arg):
    print("Welcome to all")
```

```python
gretting()
```

```python
def outer_twice(func):
    def inner_twice(*arg,**kewarg):
        print("Start of the function")
        func(*arg,**kewarg)
        func(*arg,**kewarg)
        print("End of the Function")
    return inner_twice
```

```python
@outer_twice

def my_func(name):
    print(f"My name is {name}")
    return f"Hi, {name}"
```

```python
x= my_func("Utpal")
```

```python
import time

def timer_outer(func):
    def timer_inner(*arg):
        start = float(time.time())
        func(*arg)
        end = float(time.time())
        return (end-start)
    return timer_inner
```

```python
@timer_outer

def my_func(items):
    sum=0
    for item in range(items):
        sum+=item
    return sum
```

```python
my_func(100000000)
```
