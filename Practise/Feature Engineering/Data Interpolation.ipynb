## Data Interpolation

### 1. Linear Interpolation

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x_new = np.linspace(1,5,10)
x_new
```

```python
y_interp = np.interp(x_new, x, y)
y_interp
```

```python
plt.scatter(x_new, y_interp)
plt.show()
```

### 2. Cubic Interploation

```python
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])
```

```python
plt.figure(figsize=(3,3))
plt.scatter(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
x_new = np.linspace(1,5,10)
x_new
```

```python
from scipy.interpolate import interp1d
f = interp1d(x, y, kind="cubic")
y_interp = f(x_new)
y_interp
```

```python
plt.figure(figsize=(3,3))
plt.scatter(x_new,y_interp)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

### 3. Polynomial Interpolation

```python
x = np.array([1,2,3,4,5])
y = np.array([1,8,9,16,25])
```

```python
p=np.polyfit(x,y,2) # 2 is degree of polynomial
```

```python
x_new = np.linspace(1,5,10)
x_new
```

```python
y_interp = np.polyval(p, x_new)
y_interp
```

```python
plt.scatter(x_new, y_interp)
plt.show()
```
