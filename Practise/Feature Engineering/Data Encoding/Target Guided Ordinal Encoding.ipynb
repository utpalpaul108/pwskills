## Target Guided Ordinal Encoding

```python
import pandas as pd
```

```python
df = pd.DataFrame({"city":["New Work", "London", "Paris", "Tokyo", "New Work", "Paris"],
                  "price":[200, 150, 300, 250, 180, 320]})
```

```python
df
```

```python
new_city = df.groupby("city")["price"].mean().to_dict()
new_city
```

```python
df["city_encoded"]=df["city"].map(new_city)
```

```python
df.head()
```

```python
# Internal Assignment
import seaborn as sns

df = sns.load_dataset("tips")
df.head()
```

```python
mean_time = df.groupby("time")["total_bill"].mean().to_dict()
mean_time
```

```python
df["time_encoded"] = df["time"].map(mean_time)
```
