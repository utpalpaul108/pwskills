## Anomaly Detection Using DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.datasets import make_circles
```

```python
# Make the Dataset
X,y = make_circles(n_samples=750, factor=0.3, noise=0.1)
```

```python
X
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
# Make Model
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.10)
```

```python
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
X
```

```python
plt.scatter(X[:,0], X[:,1], c=dbscan.labels_)
plt.show()
```

```python
outliers_index = np.where(dbscan.labels_<0)
outliers = X[outliers_index]
outliers
```
