## Local Outlier Factor Anomaly Detection

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
```

```python
# Make the dataset
X,y = make_circles(n_samples=800, factor=0.3, noise=0.1)
```

```python
X
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
# Make the model
from sklearn.neighbors import LocalOutlierFactor
localOutlier = LocalOutlierFactor(n_neighbors=5)
```

```python
labels = localOutlier.fit_predict(X)
```

```python
localOutlier.negative_outlier_factor_
```

```python
plt.scatter(X[:,0], X[:,1], c=labels)
plt.show()
```
