## Check Normal Distribution Using QQ Plot

```python
import scipy.stats as stat
import pylab
import numpy as np
```

```python
# We will check that every numric features of IRIS dataset is normally distributed or not

import seaborn as sns
df = sns.load_dataset("iris")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_data(df, feature):
    plt.figure(figsize=(15,4))
    plt.subplot(1,3,1)
    sns.histplot(df[feature], kde=True)
    plt.title("Histogram")
    plt.subplot(1,3,2)
    stat.probplot(df[feature], dist='norm', plot=pylab)
    plt.title("For Normal Distribution")
    plt.subplot(1,3,3)
    stat.probplot(df[feature], dist='uniform', plot=pylab)
    plt.title("For Uniform Distribution")
    plt.show()
```

```python
plot_data(df, 'sepal_length')
```

```python
plot_data(df, 'sepal_width')
```

```python
plot_data(df, 'petal_length')
```

```python
plot_data(df, 'petal_width')
```

```python
# Now we implent QQ plot for tips dataset
df = sns.load_dataset("tips")
df.head()
```

```python
plot_data(df, 'total_bill')
```

```python
plot_data(df, 'tip')
```
