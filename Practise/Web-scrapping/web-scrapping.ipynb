```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
# import logging
```

```python
flipcart_url = "https://www.flipkart.com/search?q=" + "iphone11"
flipcart_url
```

```python
url_client = urlopen(flipcart_url)
flipcart_page = url_client.read()
```

```python
flipcart_html = bs(flipcart_page, 'html.parser')
```

```python
product_divs = flipcart_html.find_all("div", {"class":"_1AtVbE col-12-12"})
len(product_divs)
```

```python
del product_divs[:3]
```

```python
product_urls = []

for product_div in product_divs:
    product_href = product_div.div.div.div.a["href"]
    product_urls.append("https://www.flipkart.com" + product_href)

len(product_urls)
```

```python
urlopen(product_urls[0]).read()
```

```python
requests.get(product_urls[0]).text
```

```python
single_product_html = bs(urlopen(product_urls[0]).read(), "html.parser")
```

```python
single_product_comments = single_product_html.find_all("div", {"class":"_16PBlm"})
len(single_product_comments)
```

```python
for single_product_commenter in single_product_comments:
    print(single_product_commenter.div.div.find_all("p", {"class":"_2sc7ZR _2V5EHH"})[0].text)
```
