## Univariate, Bivariate and Multi-Variate Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
df.species.unique()
```

df.shape

## **Univariate Analysis**

```python
df_setosa = df.loc[df['species'] == 'setosa']
df_virginica = df.loc[df['species'] == 'virginica']
df_versicolor = df.loc[df['species'] == 'versicolor']
```

```python
plt.plot(df_setosa['sepal_length'], np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'], np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'], np.zeros_like(df_setosa['sepal_length']),'o')
plt.xlabel("Sepal Length")
```

## **Bivariate Analysis**

```python
sns.FacetGrid(df, hue="species").map(plt.scatter, "petal_length", "sepal_width").add_legend()
plt.show()
```

## **Multivariate Analysis**

```python
sns.pairplot(df, hue="species", height=3)
```
