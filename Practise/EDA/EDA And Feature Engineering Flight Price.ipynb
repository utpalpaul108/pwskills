### **Feature Engineering for Duration Column**
- We take hours and minutes in numerical form

```python
# Custom function for Hours
def split_hours(duration):
    dr_split = duration.split(" ")
    if len(dr_split)<=1:
        if "h" in dr_split[0]:
            return dr_split[0].split("h")[0]
        else:
            return 0
    else:
        return dr_split[0].split("h")[0]
```

```python
# Custom function for Minutes

def split_minutes(duration):
    dr_split = duration.split(" ")
    if len(dr_split)<=1:
        if "m" in dr_split[0]:
            return dr_split[0].split("m")[0]
        else:
            return 0
    else:
        return dr_split[1].split("m")[0]
```

```python
df['Duration_hour'] = df['Duration'].map(lambda x:split_hours(x))
df['Duration_minute'] = df['Duration'].map(lambda x:split_minutes(x))
```

```python
df['Duration_hour'] = df['Duration_hour'].astype(int)
df['Duration_minute'] = df['Duration_minute'].astype(int)
```

```python
df.drop("Duration", axis=1, inplace=True)
```
